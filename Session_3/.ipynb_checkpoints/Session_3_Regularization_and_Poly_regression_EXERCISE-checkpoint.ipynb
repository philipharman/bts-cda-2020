### Import libraries and dataset from sklearn

````python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import matplotlib
matplotlib.rcParams.update({'font.size': 12})
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import metrics
from sklearn.metrics import mean_squared_error
import seaborn as sns
sns.set()

boston=load_boston()
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)

boston_df.head()
````

### Add Noise to original dataset

````python
n_houses = boston_df.shape[0]
rand_noise = np.random.rand(n_houses, 5)

rand_noise_df = pd.DataFrame(data=rand_noise, columns=['Noise_1', 'Noise_2', 'Noise_3', 'Noise_4', 'Noise_5'])
X = pd.concat([boston_df, rand_noise_df], axis=1)
y = boston['target']
````

````python
X.head()
````

````python
y[0:5]
````

### Implement classical linear regression, Lasso, Ridge and ElasticNet (use the default parameter of sklearn) and describe how the scores of the features change

````python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

# CREATE MODEL INSTANCES
lin_reg, lasso_reg, ridge_reg, eNet_reg = LinearRegression(), Lasso(), Ridge(), ElasticNet()
models = [lin_reg, lasso_reg, ridge_reg, eNet_reg]
````

##### NORMALIZE DATA USING THE ```StandardScaler``` PROVIDED BY sklearn

````python
from sklearn.preprocessing import StandardScaler

boston_ss = StandardScaler()
X_scaled = boston_ss.fit_transform(X=X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=235)
````

````python
## FITTING OF THE MODELS AND RESULTS

def plotter(model):
    sns.barplot(x = model.coef_, y = X.columns, color = 'b')
    plt.title(str(model)[:-2])
    plt.xlabel('Coefficient')
    plt.ylabel('Feature')
    plt.show()

for model in models:
    model.fit(X_train, y_train)
    plotter(model)
````

````python
results_df = pd.DataFrame(columns = ['Feature'], data = X.columns)
for model in models:
    coefs = pd.DataFrame(columns = [str(model)[:-2]], data = model.coef_)
    results_df = pd.concat([results_df, coefs], axis = 1)

fig, ax = plt.subplots(figsize=(6,8))
sns.heatmap(data = results_df.drop(columns = 'Feature'), cmap = 'vlag_r', 
            center = 0, annot = True, cbar = False, 
            yticklabels = results_df.Feature).set_ylim(len(results_df),-.05)
````

#### Decribe the results

- Linear & Ridge regression were unable to completely rule out the noise data (as expected).
- Lasso & Elastic Net successfuly set the Noise features to 0, as well as a few other features. There are also some instances where Lasso set features to 0, but Elastic net set them to very small non-zero coefficients.
- There are several cases where Linear and Ridge models set relatively high magnitude feature coefficients, but Lasso and Elastic set at/near 0 (ex. DIS, RAD, TAX).

### Compute the Model Scores

````python
for model in models:
    y_pred = model.predict(X_test)
    mse = round(metrics.mean_squared_error(y_test, y_pred),2)
    modname = str(model)[:-2]
    print(modname + ' Mean Squared Error score: ' + str(mse))
````

This might seem counter-intuitive, but it is important to remember the ridge regression model traded some variance for bias, which ultimately lead to an overall smaller error. The Lasso and Elastic Net models traded a significant amount of variance for bias, and we see that our error has increased.
Interestingly, Lasso and Elastic Net had a higher MSE than Linear Regression. But does that mean that these models are unequivocally worse? I would argue not, as the Lasso and Elastic Net models also performed feature selection, which gives us better interpretability of the models. Coefficients are interpreted as the change in dependent variable with a one unit increase in predictor value, with all other predictors held constant. In the case of complex models, the assumption of holding all other predictors constant cannot reasonably be met.

### Alpha parameter:  Apply Lasso with different values of alpha and describe how the parameters scores change


- Complete the function and apply it to the following values: [0, .0001, .5, 10]
- Describe the results

````python
names = X.columns
alphas = [0, .0001, .5, 10]

# Create a function called lasso,
def lasso(alphas):
    # Create an empty data frame
    df = pd.DataFrame()
    
    # Create a column of feature names
    df['Feature Name'] = names
    
    # For each alpha value in the list of alpha values,
    for alpha in alphas:
        # Create a lasso regression with that alpha value,
        lasso = Lasso(alpha = alpha)
        
        # Fit the lasso regression
        lasso.fit(X_scaled, y)
        
        # Create a column name for that alpha value
        column_name = 'Alpha = %f' % alpha

        # Create a column of coefficient values
        df[column_name] = lasso.coef_
        
    # Return the dataframe  
    return df
````

````python
df = lasso([0, .0001, .5, 10])

fig, ax = plt.subplots(figsize=(6,8))
sns.heatmap(data = df.drop(columns = 'Feature Name'), cmap = 'vlag_r', 
            center = 0, annot = True, cbar = False, 
            yticklabels = df['Feature Name']).set_ylim(len(df),-.05)
plt.title('Lasso Regression with Varied Alpha Values')
````

#### Describe the above results
- In general, the coefficients decrease as Alpha increases, to the extent that Alpha = 10 sets all coefficients to 0 (positing that no coefficients have any predictive power)
- Interestingly, RM actually had a higher magnitude coefficient for Alpha = 0.5 than Alpha = 0 or Alpha = 0.0001.
- There's very little difference between Alpha = 0 and Alpha = 0.0001. Alpha = 0.5 yields the simplest model, excluding Alpha = 10 (which disregards all features).

### Apply Ridge with different values of alpha and describe how the parameters scores change
- Complete the function and apply it to the following values: [0, .0001, .5, 10]
- Describe the results

````python
names = X.columns

# Create a function called lasso,
def ridge(alphas):

    # Create an empty data frame
    df = pd.DataFrame()
    
    # Create a column of feature names
    df['Feature Name'] = names
    
    # For each alpha value in the list of alpha values,
    for alpha in alphas:
        # Create a Ridge regression with that alpha value,
        ridge = Ridge(alpha = alpha)
        
        # Fit the lasso regression
        ridge.fit(X_scaled, y)
        
        # Create a column name for that alpha value
        column_name = 'Alpha = %f' % alpha

        # Create a column of coefficient values
        df[column_name] = ridge.coef_
        
    # Return the dataframe  
    return df
````

````python
df = ridge([0, .00001, .5, 10])

fig, ax = plt.subplots(figsize=(6,8))
sns.heatmap(data = df.drop(columns = 'Feature Name'), cmap = 'vlag_r', 
            center = 0, annot = True, cbar = False, 
            yticklabels = df['Feature Name']).set_ylim(len(df),-.05)
plt.title('Ridge Regression with Varied Alpha Values')
````

#### Describe the above results
- As seen previously, Ridge regression did not set any features (noise or otherwise) to zero. This held true even for Alpha = 10.
- Most feature coefficients changed very little in response to varied alpha values.

In practice, we will not manually look for the best alpha coefficient. Both in Lasso and ridge regression, the alpha is a hyperparameter. Therefore, we will find the best one via crossvalidation. Fortunately, sklearn has a function that performs CV to find the optimal alpha automatically. 

````python
from sklearn.linear_model import LassoCV

reg = LassoCV(max_iter = 5000, cv = 20, normalize=True).fit(X_train, y_train)
reg.alpha_
````
